```python
import dotenv
import requests
import os
import json
from io import StringIO
from dotenv import load_dotenv
import base64
from requests import post, get
load_dotenv()
client_id = os.getenv("CLIENT_ID")
client_secret = os.getenv("CLIENT_SECRET")
```

```python
def get_token():
    auth_string = client_id + ":" + client_secret
    auth_bytes = auth_string.encode("utf-8")
    auth_base64 = str(base64.b64encode(auth_bytes),"utf-8")

    url = "https://accounts.spotify.com/api/token"
    headers = {"Authorization": "Basic " + auth_base64,
               "Content-Type": "application/x-www-form-urlencoded"
               }
    data = {"grant_type": "client_credentials"}
    result = post(url, headers=headers, data=data)
    json_result = json.loads(result.content)
    token = json_result["access_token"]
    return token
def get_auth_header(token):
    return {"Authorization": "Bearer " + token}

def search_for_podcast(token, podcast_name):
    url = "https://api.spotify.com/v1/search"
    headers = get_auth_header(token)
    query = f"?query={podcast_name}&type=show,episode&limit=50&market=BR"

    query_url = url + query
    result = get(query_url, headers=headers)
    json_result = json.loads(result.content)
    #dicionário com as key-values correspondentes 
    json_result = json.loads(result.content)
# Dicionário com as key-values correspondentes
json_result = [{'name': item['name'], 'description': item['description'], 'id': item['id'], 'total_episodes': item['total_episodes']} for item in json_result['shows']['items']]

# Abre o arquivo com o modo "a" para adicionar novas linhas ao final do arquivo
with open('output.json', 'a') as f:
    # Converte cada item da lista em uma string JSON e adiciona um new line delimiter (\n)
    for item in json_result:
        json.dump(item, f)
        f.write('\n')

# Lê o arquivo e imprime seu conteúdo para conferência
with open('output.json', 'r') as f:
    for line in f:
        print(line.strip())

token = get_token()
search_for_podcast(token,"Data%20Hackers")
```
